```python
from PIL import Image
import mindspore as ms
import numpy as np
import network as nw
from mindspore.dataset import vision
import os
```

```python
def visualize_model(path):
    image = Image.open(path).convert("RGB")
    image = image.resize((128, 128))
    image = np.array(image)

    # 归一化处理
    image = vision.c_transforms.Rescale(1.0 / 255.0, 0)(image)

    # # 图像通道由(h, w, c)转换为(c, h, w)
    image = vision.c_transforms.HWC2CHW()(image)

    # 扩展数据维数为(1, c, h, w)
    image = np.expand_dims(image, axis=0)
    # print(image.shape)

    # 加载网络
    net = nw.network()

    param_dict = ms.load_checkpoint("./checkpoint/MyNet.ckpt")
    ms.load_param_into_net(net, param_dict)
    model = ms.Model(net)

    # 模型预测
    pre = model.predict(ms.Tensor(image))
    result = np.argmax(pre)
    return result
```

```python
if __name__ == '__main__':
    i=0
    a=0
    while 1:
        if os.path.exists("./mindcon_food_classification/test/img_"+str(i)+".jpg"):
            print(visualize_model("./mindcon_food_classification/test/img_"+str(i)+".jpg"))
            a=0
        else:
            a+=1
            if a>10:
                break
        i+=1
```
